```python
def markdown_parser(markdown):
    
#find out if there is substring surrounded by underscores
    
    x = re.search("\_(.*?)\_", markdown)
    

    if x:
        us_string = x.group()    
        
# If it does, find out if there is a space in the substring 

        space = " "
        
        pos_space = [pos for pos, char in enumerate(us_string) if char == space]
            
# if there is no space, append the html emphasis...

        if not pos_space:
            clean_us_string = re.sub(r"^_+|_+$", "", us_string)
            emphasised_string = (f"<em>{clean_us_string}</em>")
            first_md = re.sub(us_string, emphasised_string, markdown)

#if there is, find out the position of the first space, the last space and the length of the string

        else:
            first_space = pos_space[0]
            last_space = pos_space[-1]
            ls = len(us_string)
           
# if space is first or last in substring, no emphasis, so return original string
            
            if first_space == 1 or last_space == (ls - 2):
                first_md=markdown
                
# if space isnot first or last in substring, emphasis, append the html emphasis

            else:
                clean_us_string = re.sub(r"^_+|_+$", "", us_string)
                emphasised_string = (f"<em>{clean_us_string}</em>")
                first_md = re.sub(us_string, emphasised_string, markdown)

# if no underscore, no emphasis, so return original string                
                
    else:
        first_md = markdown
    
    print(first_md)
        
# Repeat code in case of double emphasis

    x = re.search("\_(.*?)\_", first_md)
    

    if x:
        us_string = x.group()    
        space = " "
        pos_space = [pos for pos, char in enumerate(us_string) if char == space]            
        if not pos_space:
            clean_us_string = re.sub(r"^_+|_+$", "", us_string)
            emphasised_string = (f"<em>{clean_us_string}</em>")
            second_md = re.sub(us_string, emphasised_string, first_md)
        else:
            first_space = pos_space[0]
            last_space = pos_space[-1]
            ls = len(us_string)
                
            if first_space == 1 or last_space == (ls - 2):
                second_md = first_md       
            else:
                clean_us_string = re.sub(r"^_+|_+$", "", us_string)
                emphasised_string = (f"<em>{clean_us_string}</em>")
                second_md = re.sub(us_string, emphasised_string, first_md)                
    else:
        second_md = first_md
                
# Find out if the final variable contains hashes
# If it does, split out the hashtags and the content and removing trailing and leading spaces in header content
# Then, Count the number of hashes to determine the header level
# Transform the markdown into HTML
# If it doesn't, return the final variable as is


# Step 1: Find out if string has hashes
    a = re.search("^#+ ", second_md)
    c = re.search("^ +#+ ", second_md)
    #print(a)
    #print(c)

    # Step 2: If it doesn't return the markdown
    if a or c:
    	b = re.split("#\s", second_md, 1)
    	#print(b)
    else:
    	return(second_md)
    
    # Step 3: Split out the hashtags and the content
    header_hashtags = b[0]
    raw_header_content = b[1]
    header_content = re.sub(r"^\s+|\s+$", "", raw_header_content)
    #print(header_hashtags)
    
    # Step 4: Count the number of hashes to determine the header level
    hashtag ="#"
    header_level = header_hashtags.count(hashtag)+1
    
    #print(header_level)
    
    # Step 5: Transform the markdown into HTML    
    if header_level == 1:
    	return(f"<h1>{header_content}</h1>")
    elif header_level == 2:
    	return(f"<h2>{header_content}</h2>")
    elif header_level == 3:
    	return(f"<h3>{header_content}</h3>")
    elif header_level == 4:
    	return(f"<h4>{header_content}</h4>")
    elif header_level == 5:
    	return(f"<h5>{header_content}</h5>")
    elif header_level == 6:
    	return(f"<h6>{header_content}</h6>")
    else:
    	return(second_md)
            
print(markdown_parser("#### I have _always_ been in _Love Island_"))
```
